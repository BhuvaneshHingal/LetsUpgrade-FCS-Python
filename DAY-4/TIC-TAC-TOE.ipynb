```python
a=[['_','|','_','|','_'],['_','|','_','|','_'],['_','|','_','|','_']]
b=[[0,"|",1,"|",2],[3,"|",4,"|",5],[6,"|",7,"|",8]]
c=str(input("Enter O or X as per choice: "))
if c=="X" or c=="x":
    v="O"
else:
    v="X"
n1=int(input("Enter 1 if you wish to play first, or enter 2 if computer should play first "))
print("For the prupose of the game, indices would be as ")
for i in b:
    for j in i:
        print(j,end=' ')
    print()
p1,cmp=[],[]
u,co,aa=0,0,0
import random
def win(x):
    if True:
        for i in range(0,5):
            f=1
            if a[0][i]==x:
                for j in range(1,3):
                    if a[j][i]==x:
                        f=f+1
            if f==3:
                return True
    if True:
        for i in range(0,3):
            f=0
            for j in range(0,3):
                if str(a[j][i])==x:
                    f=f+1
            if f==3:
                return True
    if True:
        f=0
        for i in range(0,3):
            if (a[i][i*2])==x:
                f=f+1
        if f==3:
            return True
    if True:
        f=0
        for i in range(0,3):
            if (a[i][4-(2*i)])==x:
                f=f+1
        if f==3:
            return True
    return False    
def tie():
    if win(c)==False and win(v)==False and count==9:
        return True
    else:
        return False
        
def index(n):
    for i in range(0,len(b)):
        for j in range(0,len(b[i])):
            if b[i][j]==n:
                return j
def count():
    c1=0
    for x in a:
        for y in x:
            if str(y)=="O" or str(y)=="X":
                c1=c1+1
    return c1
def display():
    for x in a:
        for y in x:
            print(y,end=' ')
        print()
def repeat(w):
    w=int(input("Box is already taken, reenter "))
    return w
def check():
    if win(c):
            print("User wins ")
            display()
            return 1
    elif win(v):
            print("Computer wins; try again next time ")
            display()
            return 1
    elif tie():
            print("Match has been a tie ")
            return 1
def user():
    u=int(input("Enter the index where you wish to play 0-8? "))
    while u in p1 or u in cmp:
        u=repeat(u)
    p1.append(u)
    u1=index(u)
    a[u//3][u1]=c
def comp():
    co=random.randint(0,8)
    while (co in p1 or co in cmp):
        co=random.randint(1,8)
    cmp.append(co)
    print("The computer chose the index",end=' ')
    print(co)
    co1=index(co)
    a[co//3][co1]=v
if n1==1:
    for k in range(0,9):
        if k==0 or k%2==0:
            user()
        else:
            comp()
        display()
        if check()==1:
            aa=1
            break
    if aa==0:
        aa=check()
elif n1==2:
    for k in range(0,9):
        if k%2==1:
            user()
        else:
            comp()
        display()
        if check()==1:
            aa=1
            break
    if aa==0:
        aa=check()
if win(c)==False:
    if win(v)==False:
        print("DRAW")
```
